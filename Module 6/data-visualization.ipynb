## Importance of Data Visualization

1. **Pattern Recognition:** Visualizations like charts and graphs allow viewers to easily identify trends, relationships, and outliers within data that might be missed in raw numbers. This swift grasp of patterns facilitates deeper understanding and faster communication of insights.

2. **Information Compression:** Complex datasets with numerous data points can be overwhelming in raw form. Visualization condenses information into a clear and concise format, highlighting key elements and enabling viewers to absorb the message quickly.

3. **Enhanced Engagement:** Human brains are wired to process visuals more effectively than text. Compelling charts and graphs capture attention, promote information retention, and make data analysis a more engaging experience for both the presenter and the audience.

## Introduction to Matplotlib

Matplotlib is a powerful Python library for creating static, animated, and interactive visualizations. It empowers you to transform raw data into insightful graphs and charts, simplifying data exploration and communication.

### Types of Data Visualization in Matplotlib

### 1. Line Plot

The line plot, a cornerstone of data visualization, excels at revealing trends over time or across different categories. It connects data points with a line, visually depicting how a value changes in relation to another variable.

```python
# Data for number of orders per month
data = {
    "Months": ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"],
    "No. Of Orders": [8, 10, 12, 14, 18, 20, 26, 24, 22, 14, 12, 9],
    "No. Of Orders for Jacket": [8, 10, 12, 14, 18, 20, 26, 24, 22, 14, 12, 9],
    "No. Of Orders for Swim Wear": [25, 27, 24, 20, 15, 13, 9, 13, 14, 16, 18, 22]
}

order_per_month_df = pd.DataFrame(data)
order_per_month_df
```

```python
# call matplotlib and set the axis x and y by .plot(x, y)
# plt.plot(order_per_month_df["Months"], order_per_month_df["No. Of Orders"])

# options <marker>, <color>, <linestyle>, <linewidth>
plt.plot(order_per_month_df["Months"], order_per_month_df["No. Of Orders"], marker="o", color="blue", linestyle="dashed", linewidth=3)

# adding the labels
plt.xlabel("Month")
plt.ylabel("No. of Orders")



# adding a title
plt.title("Number of Orders per Month")

# display the grid
plt.grid(True)

# display the plot
plt.show()
```

```python
# compare two product
plt.plot(order_per_month_df["Months"], order_per_month_df["No. Of Orders for Jacket"], marker="o", color="blue", linestyle="dashed", linewidth=3, label="Jacket")
plt.plot(order_per_month_df["Months"], order_per_month_df["No. Of Orders for Swim Wear"], marker="s", color="red", linestyle="dashed", linewidth=3, label="Swim Wear")

# adding the labels
plt.xlabel("Month")
plt.ylabel("No. of Orders")

# adding a title
plt.title("Number of Orders per Month")
plt.legend()  # Use the labels set in above as legend here

# display the grid
plt.grid(True)

# display the plot
plt.show()
```

### 2. Scatter Plot

The scatter plot, a versatile tool in the data visualization toolbox, excels at revealing relationships between two numerical variables. Each data point is represented by a dot, plotted on a coordinate plane according to its corresponding values. By observing the distribution of these dots, we can uncover hidden patterns and potential correlations.

```python
# Data for Ten Year Weight Record
data = {
    "Years": [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015],
    "Weights": [80, 82, 84, 85, 87, 87, 84, 80, 79, 76]
}

weight_df = pd.DataFrame(data)
weight_df
```

```python
# Create a scatter plot
plt.scatter(weight_df["Years"], weight_df["Weights"])

# add labels
plt.xlabel("Years")
plt.ylabel("Weights (kg)")

# add a grid
plt.grid(True)

# add a title
plt.title("Changes in Weights over Ten Years")

# display a plot
plt.show()
```

```python
# Customize the plot

plt.scatter(weight_df["Years"], weight_df["Weights"], s=80, color="green", alpha=.6, marker="s")

# add labels
plt.xlabel("Years")
plt.ylabel("Weights (kg)")

# add a grid
plt.grid(True)

# add a title
plt.title("Changes in Weights over Ten Years")

# display a plot
plt.show()
```

### 3. Pie Charts

Pie charts, those delightful circles sliced into wedges, are a go-to choice for visualizing proportions of a whole.  Matplotlib's pie chart functionality, accessible through the pyplot.pie() function, empowers you to represent data categories and their relative sizes in a clear and intuitive way.

```python
# Data for spending amounts for each category
data = {
    "Categories": ['Rent', 'Food', 'Transportation', 'Savings', 'Other'],
    "Amounts": [500, 300, 200, 100, 100]
}

spending_amount_df = pd.DataFrame(data)
spending_amount_df
```

```python
# Create a Pie Chart
# autopct:'%.1f%%' - to 1 decimal point, '%.2f%%' - to 2 decimal point
plt.pie(spending_amount_df["Amounts"], labels=spending_amount_df["Categories"], autopct='%.0f%%')

plt.title("Budget for each Category")

plt.show()
```

```python
# Customization

# colors: ensure the number of colors match the number of data(categories here).
# colors: can be color name, hexadecimal(#ffffff), RGB(0-1), RGBA(0-1)
# pctdistance: location of % (distance from the centre)
plt.pie(spending_amount_df["Amounts"], labels=spending_amount_df["Categories"], autopct='%.0f%%',
        colors=["red", "green", "#0000ff", (0, 1, 1), (1, 0, 1, 0.3)], pctdistance=0.7)

plt.title("Budget for each Category")

# loc: location of the legend
plt.legend(labels=spending_amount_df["Categories"], title="Categories", loc="upper left")

plt.show()
```

#### 4. Bar Charts

Bar charts, the workhorses of data visualization, excel at showcasing comparisons between different categories.  Matplotlib's bar chart functionality, primarily accessed through the pyplot.bar() function, allows you to translate numerical data into a series of rectangular bars. The height of each bar corresponds to the value it represents, while the bars are positioned along a horizontal axis that typically represents the categories being compared.

```python
# Data for student enrollment per course
data = {
    "Course": ['Data Analytics', 'Frontend Development', 'DevOps', 'Full-Stack Development'],
    "Students": [30, 25, 32, 28]
}

# Create a data frame
students_per_course_df = pd.DataFrame(data)
students_per_course_df
```

```python
# Create a Bar Chart
plt.bar(students_per_course_df["Course"], students_per_course_df["Students"])

# Add labels
plt.xlabel("Courses")
plt.ylabel("No. of Students")

# Add a title
plt.title("Students per Course")

# lotate the label
plt.xticks(rotation=30)

# show the plot
plt.show()
```

```python
# Customize the chart

plt.bar(students_per_course_df["Course"], students_per_course_df["Students"], color=["green", "violet", "purple", "orange"])

# Add labels
plt.xlabel("Courses", color="blue")
plt.ylabel("No. of Students", color="blue")

# Add a title
plt.title("Students per Course")

# lotate the label
plt.xticks(rotation=30)

# show the plot
plt.show()
```

```python
# Create the Bar Chart

plt.bar(students_per_course_df["Course"], students_per_course_df["Students"], 

        color=["blue", "red", "green", "violet"])

# Set the label for x and y axis

plt.xlabel("Courses")

plt.ylabel("No. Of Students")

# Add the title

plt.title("Students enrolled per Course", fontsize=16, fontweight="bold")

# Rotate the label for x-axis by 45 degree

plt.xticks(rotation=45)

# Show the Bar Chart

plt.show()




```

```python
# Create the Bar Chart

plt.bar(students_per_course_df["Course"], students_per_course_df["Students"], 

        color=["blue", "red", "green", "violet"], width=0.9, edgecolor="red")

# Set the label for x and y axis

plt.xlabel("Courses")

plt.ylabel("No. Of Students")

# Add the title

plt.title("Students enrolled per Course", fontsize=16, fontweight="bold")

# Rotate the label for x-axis by 45 degree

# plt.xticks(rotation=45)

# Show the Bar Chart

plt.show()




```
